```python
# Import lớp Image
from PIL import Image
from IPython.display import display #để hiển thị ra hình ảnh
from imgtools import *
```

```python
# Đường dẫn tới hình ảnh
my_path = "dfe7e494-bba9-46a4-bd48-f032dbc2570f.png" #Đường dẫn đến file hình ảnh
```

```python
# Đọc ảnh 
img = load_image(my_path)
```

```python
display(img)
```

```python
# Kích thước ban đầu:
print(img.size)
```

```python
# Thay đổi kích thước:
new_size = (500,500)
# Tạo hình ảnh mới:
new_image = img.resize(new_size) #resize thay đổi kích thước không làm thay đổi ảnh cũ.
display(new_image)
```

```python
# Task: Lưu hình ảnh thành một file mới
thumbnail_size = (50,50)
# Tạo hình ảnh mới:
new_image_2 = img.thumbnail(thumbnail_size) #thumbnail thay đổi làm biến dạng ảnh cũ.
display(new_image_2) #Giải thích: in ra None vì ảnh này bị đè bởi ảnh trước và bị biến dạng nên ko ra ảnh, nêu dùng giá trị ảnh trước là img => sẽ ra kết quả! 
```
